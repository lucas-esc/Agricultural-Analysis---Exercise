Hi. This is an exercise that I did to train data analysis using Pandas. 

In the first part I plotted two graphs showing how the planting of potatoes in Brazil has grown while the area used for cultivation has decreased over the years. This is possible thanks to innovations in the agriculture sector.

In the second part I generated a data frame and graph showing the total capital from agriculture of the top ten producing countries and their progress from 1961 to 2019.

#### I'm using a csv file disponibilized by FAOSTAT here: https://www.fao.org/faostat/en/#data

```python
# First, we need to import pandas and set it to read our file

import pandas as pd
all_data = pd.read_csv("../input/faostat-production-crops-livestock-e-all-data/Production_Crops_Livestock_E_All_Data_(Normalized).csv", encoding='cp1252')
all_data
```

```python
#Here I create a data frame specific for an area searching setting the new data frame like this: 

brazil_all_data = all_data[all_data['Area'] == 'Brazil']
brazil_all_data
```

```python
# The .dropna method is usefull for with it you can clean your data frame from empty values, but here I will not use it

#brazil_all_data.dropna()
```

```python
# Again, we now filter our data frame to show only the series related to the item "Potatoes" 

brazil_potatoes = brazil_all_data[brazil_all_data['Item'] == 'Potatoes']
brazil_potatoes
```

```python
# And again whith only the series that have the Element "Production" 

brazil_potatoes_prod =  brazil_potatoes[brazil_potatoes['Element'] == 'Production']
brazil_potatoes_prod
```

### Now we will use the MatPlotLib to plot our data frame into a graphic for best visualization

```python
# First, we import the Matplot function lyke this:

import matplotlib.pyplot as plt



plt.rcParams["figure.figsize"] = [20, 15] #Here I setted the size of every graphic gerated to this figsize


# Now we plot our data frame into a graphic and set the parameters

brazil_potatoes_prod.plot.bar(x='Year', y='Value', ylabel='Milhoes de Toneladas', xlabel='Ano', grid=True, 
                              title="Produção de Batatas no Brasil")
```

```python
# We will do the same here, but now creating a data frame to "Area harvested"

brazil_potatoes_hec = brazil_potatoes[brazil_potatoes['Element'] == 'Area harvested']
brazil_potatoes_hec
```

```python
# And plot with the parameters that we want to show in our graphic

brazil_potatoes_hec.plot.bar(x='Year', y='Value', ylabel='Milhares de Hectares', xlabel='Ano', grid=True, 
                              title="Área Plantada de Batatas no Brasil")
```

```python
# Okay, so we have the two graphics, but we wanto them to be ploted together. So first we need to merge the two data frames

brazil_potatoes_merged = pd.merge(brazil_potatoes_prod, brazil_potatoes_hec, on = ['Area Code', 'Area', 'Item Code',
                                                                                   'Item', 'Year', 'Year Code'])

# Here I renamed some of the column names. So when we plot, the graphic will have these names 

brazil_potatoes_merged_renamed = brazil_potatoes_merged.rename(columns={'Value_x':'Produção', 'Value_y':'Hectares'})

brazil_potatoes_merged_renamed
```

```python
# Now we can plot the merged data frames graphic. As we are ploting two x-velues, we need to set the secondary y-value too
# if we dont do this, our graphic scale will be too hight in "Production" y-value

graphic = brazil_potatoes_merged_renamed.plot.bar(x='Year', y=['Produção', 'Hectares'], secondary_y='Produção', xlabel='Ano', 
                                legend=True, title='Área Plantada X Produção de Batatas no Brasil')

# Here we just define the labels of our graphic

graphic.set_ylabel('Milhares de Hectares')
graphic.right_ax.set_ylabel('Milhões de Toneladas')
```

### Now we will try to see how the amount of all gross production of all world improved a long the years. 

```python
# First we open our csv file

all_value_data = pd.read_csv("../input/faostat-value-of-production-e-all-data/Value_of_Production_E_All_Data_(Normalized).csv", encoding='cp1252')

# For this exercise we can delete some of the columns to turn our data frame in a small and clean 

del all_value_data['Area Code'] 
del all_value_data['Item Code'] 
del all_value_data['Element Code'] 
del all_value_data['Year Code']
all_value_data
```

```python
# Here we can see wich currency our data frame has using .value_counts that stack all the repeated values in a column 

all_value_data['Element'].value_counts()
```

```python
# We will use the "Gross Production Value (constant 2014-2016 thousand I$)"

all_value_data_grossp = all_value_data.loc[all_value_data['Element']=='Gross Production Value (constant 2014-2016 thousand I$)']
all_value_data_grossp
```

```python
# The data frame has values to áreas, regions and countries. We only want the countries for this exercice, so we select
# the Flag "Fc"

all_value_data_grossp1 = all_value_data_grossp.loc[all_value_data_grossp['Flag'] == 'Fc']

all_value_data_grossp1
```

```python
all_value_data_grossp1.columns = ['Área', 'Item', 'Produção', 'Ano', 'Moeda', 'Valor', 'Flag']
all_value_data_grossp1
```

```python
all_value_data_grossp1['Área'].value_counts()
```

```python
# Now we only want the usefull information to plot our graphic. So we create a new data frame with them.

value_data_new = all_value_data_grossp1[['Área','Ano','Valor']]
value_data_new
```

```python
# So we have now a data frame with many countrys and the years in the columns. But we want to see the only ten more productive 
# countrys. To do this, we can re-organize our data frame like this:

value_data_new = value_data_new.pivot_table(index='Área', columns='Ano', values='Valor', aggfunc='sum')
value_data_new
```

```python
# Now, in order to get the 10 more productive countrys, we need to organize our data from the total that each country produced.
# First we create a new column and set it to be the sum of all production a long the years. 

value_data_new['Total'] = value_data_new.sum(axis=1)
data_decrescent = value_data_new.sort_values(['Total'], ascending=False, axis=0)

# Now we slice the first 10

top10 = data_decrescent.head(10)
top10
```

```python
# We can alternate the columns and index rows using .transpose for better visualization

top10_df = top10.transpose()
top10_df
```

```python
# We fill the NaN values with 0 and drop the "Total" row that we created before. If you dont do this, your area graphic will
# be out of scale

top10_df.fillna(0, inplace=True)
top10_final = top10_df.drop(labels='Total', axis=0)
```

```python
# Finally, we can plot our data frame into a line graphic!

ax = top10_final.plot(kind='line', figsize=(20,8), stacked=False)
ax.set_title('Produção Bruta das Principais Nações de 1961 à 2019', fontsize=20, fontweight='bold')
ax.set_ylabel('Intervalo em Bilhoes (Int $)', fontsize=15)
ax.set_xlabel('')
ax.tick_params(labelsize=15)
```
